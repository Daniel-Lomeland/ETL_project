```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
county_file = "../Data/county_list.csv"
```

```python
county_df = pd.read_csv(county_file)
county_df.head()
```

```python
cols = [" County Name ", " County # "]
county_clean = county_df[cols]
cols_to_rename = {" County Name ": "county_name", 
                  " County # ": "county_num"}



county_clean = county_clean.rename(columns = cols_to_rename)
county_clean = county_clean.set_index("county_num")
county_clean.head()
```

```python
county_clean.to_csv("../Data/cleaned_county_list.csv", index = True, header = True)
```
