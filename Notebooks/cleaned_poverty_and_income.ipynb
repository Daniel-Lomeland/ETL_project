```python
import pandas as pd
from sqlalchemy import create_engine
import pymysql
pymysql.install_as_MySQLdb()
```

```python
# Load CSV File

df= pd.read_csv('../Data/income_&_poverty_2017.csv')

df.head()
```

```python
# Check Completeness of Records
df.count()
```

```python
# Check Data Types
df.dtypes
```

```python
# Create appropriate Column Headers

headers = {'4080377': 'pop_in_poverty',
            '14.7':"%_pop_in_poverty",
            '454312':'county',
            '59195': 'median_income',
            '58712': 'lower_bound_income',
            '59678': 'upper_bound_income'
             }
# Save DataFrame with new headers

headers_df= df.rename(columns = headers)


# Create new DataFrame with columns to keep.

clean_df = headers_df.loc[:, ["county","pop_in_poverty","%_pop_in_poverty", "median_income",
                                    "lower_bound_income", "upper_bound_income"]]





```

```python
clean_df.head()
```

```python
# Create CSV of Cleaned Data

clean_df.to_csv("../Data/cleaned_income_&_poverty_2017.csv", index = True, header = True)

engine = create_engine(f'mysql://root:PASSWORD@localhost/etl_project')

clean_df.to_sql(name = 'tax_info', con = engine, if_exists = 'append', index = True)
```
